```python
import numpy as np
import matplotlib.pyplot as plt
```

```python

def gaussian(x, mu, sig):
    return np.exp(-np.power(x - mu, 2.) / (2 * sig**2.))
def gaussian2D(x, y, mu, sig):
    [X, Y] = np.meshgrid(x, y)
    return np.dot(gaussian(x, mu[0], sig[0])[None].T, gaussian(y, mu[1], sig[1])[None])#np.exp(-((X - mu[0])**2 / (2 * sig[0]**2.) + (Y-mu[1])/ (2 * sig[1]**2.)) )
```

```python
mu0 = 0
sigma = 0.5
mu1 = mu0+np.random.rand(1)*1.22*sigma

lim = 10
n = 200
x = np.linspace(-lim,lim, n)

G1 = gaussian2D(x, np.copy(x), (0,0), (0.5,0.5))
print(G1)
```

```python
plt.imshow(G1)
plt.show()
```

```python
center = [0,0]
sigma = 1
nyquist = sigma/2.

positions = np.random.randn(4,2)*nyquist

pic = 0
for p in positions:
    pic += np.random.rand(1)*gaussian2D(x, np.copy(x), p, (sigma, sigma))
    
plt.subplot(121)
plt.imshow(gaussian2D(x, np.copy(x), [0,0], (sigma, sigma)), cmap =  'gist_stern')
plt.axis('off')
plt.subplot(122)
plt.imshow(np.array(pic), cmap  = 'gist_stern')
plt.axis('off')
plt.show()
```
